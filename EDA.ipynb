```python
#Step 1: Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")  
%matplotlib inline  
```

```python
# Reading the csv file using the read_csv function in pandas and then viewing the content with head function
df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

```python
# Pandas has a feature to find the shape/dimensions of the dataset. The below code shows how it's done:
# We use the isnull() to find all the null spaces in the datset.
print("Shape of dataset:", df.shape)
print("\nInfo:")
df.info()

print("\nMissing values:")
print(df.isnull().sum())
```

```python
# Using describe() we can see mean,standard deviation, count, min, max
df.describe()
```

```python
df.hist(figsize=(12,8), bins=20)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df.select_dtypes(include= 'number'))
plt.title("Boxplot of Numeric Features")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()

plt.figure(figsize=(8, 5))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include='number')

# Compute and plot the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Numeric Features")
plt.show()


```

```python
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived')
plt.show()
```

```python
numeric_cols = df.select_dtypes(include='number')
print("Skewness of numeric features:")
print(numeric_cols.skew())
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Age', hue='Survived', bins=30, kde=True)
plt.title("Age Distribution by Survival")
plt.show()
```
